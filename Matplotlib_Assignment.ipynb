Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of
Matplotlib.

Matplotlib is a Python data visualization library used to create static, animated, and interactive visualizations in Python programming language. It provides a wide range of tools for constructing visualizations in Python, including bar plots, line plots, scatter plots, histograms, pie charts, and more.

Matplotlib is used to represent data graphically, allowing for better understanding and analysis of complex data. With Matplotlib, users can create high-quality visualizations of various types of data, from simple line charts to complex 3D plots.

The Pyplot module of Matplotlib is a sub-library that provides a convenient interface for creating plots and charts. Here are five plots that can be plotted using the Pyplot module of Matplotlib:

1. Line Plot: A line plot is a graph that displays data as a series of points connected by straight lines. This plot is useful for showing the trend of a variable over time.

2. Scatter Plot: A scatter plot is a graph that displays data as a collection of points. This plot is useful for showing the relationship between two variables.

3. Bar Plot: A bar plot is a graph that displays data as rectangular bars. This plot is useful for comparing the quantities of different variables.

4. Histogram: A histogram is a graph that displays data as a series of bars. This plot is useful for showing the distribution of a variable.

5. Pie Chart: A pie chart is a graph that displays data as a circle divided into slices. This plot is useful for showing the proportions of different variables.

Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

# Note: Also add title, xlabel, and ylabel to the plot.
```

A scatter plot is a type of graph used to visualize the relationship between two continuous variables. It is created by plotting one variable on the x-axis and the other variable on the y-axis, and then using dots to represent the individual data points.

The main purpose of a scatter plot is to identify whether there is a relationship between the two variables and, if so, to determine the nature of that relationship. If the dots on the scatter plot form a recognizable pattern, such as a straight line or a curve, then there is likely a relationship between the variables.

Scatter plots can also be used to identify outliers or clusters of data points, which may be important in understanding the underlying data. Additionally, scatter plots can be enhanced by adding additional information, such as color or size, to represent a third variable, making it possible to explore relationships among three variables.


```python
plt.scatter(x,y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Random Data")
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

A bar plot is a type of chart that displays rectangular bars with lengths proportional to the values that they represent. The bars can be plotted horizontally or vertically, with the height or width of the bar indicating the value of the variable being plotted. Bar plots are commonly used to display and compare the relative magnitudes of discrete categories.

Bar plots are useful for visualizing categorical data, and are commonly used in market research, public opinion polling, and social science research. They can be used to display counts, proportions, frequencies, and other types of data.

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
plt.bar(company, profit)
plt.title("Bar Plot")
plt.xlabel("Company")
plt.ylabel("Profit")
plt.show()
```

```python
# bar-h(h=horizontal)
plt.barh(company, profit)
plt.title("Horizontal Bar Plot")
plt.xlabel("Profit")
plt.ylabel("Company")
plt.show()
```

Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
data = [box1, box2]
plt.boxplot(data)
plt.title("Box Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```
